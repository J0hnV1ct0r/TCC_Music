```python

# Sua chave da API do YouTube
def get_video_upload_date(video_url):
    headers = {"User-Agent": "Mozilla/5.0"}
    response = requests.get(video_url, headers=headers)

    if response.status_code != 200:
        return None  # Retorna None se não conseguir acessar o site

    soup = BeautifulSoup(response.text, "html.parser")

    # O padrão da data pode mudar, então usamos regex para extrair a data
    match = re.search(r'(?i)uploadDate":"(\d{4}-\d{2}-\d{2})"', response.text)
    if match:
        return match.group(1)  # Retorna a data no formato YYYY-MM-DD

    return None
# Exemplo com um dataset
#df = pd.DataFrame({"video_url": ["https://www.youtube.com/watch?v=3JZ_D3ELwOQ"]})
spotify_youtube["upload_date"] = spotify_youtube["Url_youtube"].apply(get_video_upload_date)

```

```python
# Sua chave da API do YouTube
API_KEY = ""

# Inicializar a API
youtube = build("youtube", "v3", developerKey=API_KEY)

def get_video_date(video_url):
    # Extrair ID do vídeo do link
    video_id = re.search(r"v=([a-zA-Z0-9_-]+)", video_url)
    if video_id:
        video_id = video_id.group(1)
    else:
        return None  # Se não encontrar um ID válido

    # Buscar detalhes do vídeo
    request = youtube.videos().list(
        part="snippet",
        id=video_id
    )
    response = request.execute()

    if "items" in response and len(response["items"]) > 0:
        return response["items"][0]["snippet"]["publishedAt"]  # Data de publicação
    return None

# Exemplo com um dataset
#df = pd.DataFrame({"video_url": ["https://www.youtube.com/watch?v=3JZ_D3ELwOQ"]})
spotify_youtube_100["upload_date"] = spotify_youtube_100["Url_youtube"].apply(get_video_date)
```
